
3. Explain the following terms

 The difference in np.array, np.asarray and np.asanyarray
 


| Function          | Description                                                           | Copies data?                                                                      | Preserves subclass (e.g., matrix)? |
| ----------------- | --------------------------------------------------------------------- | --------------------------------------------------------------------------------- | ---------------------------------- |
| `np.array()`      | Creates a new NumPy array from any object.                            | **Always** makes a copy by default unless `copy=False` is explicitly set and safe |                                    |
| `np.asarray()`    | Converts input to array **without copying** if it's already an array. | Avoids copy if not needed                                                         | **No** (returns `ndarray` always)  |
| `np.asanyarray()` | Like `np.asarray`, but **preserves subclass** (e.g., `np.matrix`).    | Avoids copy if not needed                                                         | **Yes** (keeps matrix subclass)    |


```python
import numpy as np

lst = [1,2,3]

arr1 = np.array(lst) # Always return ndarray
arr2 = np.asarray(lst) # Returns ndarray, avoids copy if input is already ndarray
arr3 = np.asanyarray(lst) # Returns ndarray or matrix, depending on input
```

```python
# Example with subclass

mat = np.matrix([[1,2],[3,4]])

print(type(np.asarray(mat)))
print(type(np.asanyarray(mat)))
```

 The difference between Deep copy and shallow copy

| Type             | What it does                                                       | Independent?          |
| ---------------- | ------------------------------------------------------------------ | --------------------- |
| **Shallow Copy** | Creates a new object, but references the **same underlying data**. | ❌ Changes affect both |
| **Deep Copy**    | Creates a new object **and** a full copy of the data inside.       | ✅ Completely separate |


```python
import numpy as np

original = np.array([1,2,3])

# swallow copy (view)
shallow = original.view()
shallow[0] = 100
print(original)

# Deep copy
deep = original.copy()
deep[0] = 200
print(original)
print(deep)
```

5.  Create a NumPy array with random integers between 1 and 10 of shape (5, 6). After creating the array 
perform the following operations:

 a)Extract all even integers from array.
 
 b)Extract all odd integers from array

```python
import numpy as np

# Step 1: Create the array
array = np.random.randint(1,11, size=(5,6))

# Step 2a: Extract even integers
even_numbers = array[array % 2 == 0]

# Step 2b: Extract odd integers
odd_numbers = array[array != 0]

#Print Result
print("Original Array:\n" ,array)
print('\nEven Numbers:\n' , even_numbers)
print('\nOdd Numbers:\n ', odd_numbers)
```

6.  Create a 3D NumPy array of shape (3, 3, 3) containing random integers between 1 and 10. Perform the 
following operations:

 a) Find the indices of the maximum values along each depth level (third axis).
 
 b) Perform element-wise multiplication of between both array

```python
import numpy as np

#Step1: Create a 3d Array
array = np.random.randint(1,11, size=(3,3,3))

#Step 2a: Indices of max values along the depth axis (axis=2)
max_indices = np.argmax(array,axis=2)

# Step 2b: Element-wise multiplication (array * array)
multiplied_array = array * array

print("Original 3d array:", array)
print("\nIndices of Max Values along Depth Axis:\n", max_indices)
print("\nElement-wise Squared array\n", multiplied_array)
```

7.  Clean and transform the 'Phone' column in the sample dataset to remove non-numeric characters and 
convert it to a numeric data type. Also display the table attributes and data types of each column

```python
import pandas as pd
import re

df = pd.read_csv(r"C:\Users\udit2\Documents\GitHub\Assignment\Resources\People Data.csv")

# Step 1: Clean 'Phone' column (remove non-numeric characters)
df['Phone'] = df['Phone'].astype(str).str.replace(r'\D','',regex=True)

# Step 2: Filter valid length
df['Phone'] = df['Phone'].apply(lambda x: x if 10 <= len(x) <= 13 else None)

# Step 3: Convert to numeric data type
df['Phone'] = pd.to_numeric(df['Phone'],errors='coerce')

# Step 4: Disable scientific notation display
pd.set_option('display.float_format',lambda x: '%.0f' % x)


# Step 5: Display table attributes and data types
print("Cleaned Phone Column:")
print(df['Phone'])

print("\n Data Types of each column:")
print(df.dtypes)
```

14. Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and 
Perform the following tasks using Matplotlib:

a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and  right y-axis for 'Humidity').

b) Label the x-axis as 'Date'.

c) Set the title of the plot as 'Temperature and Humidity Over Time'.

```python
# Step 1: Create a time-series DataFrame
date_range = pd.date_range(start='25-07-01',periods=30,freq='D')
temperature = np.random.uniform(low=25,high=40,size=30)
humidity = np.random.uniform(low=40,high=90,size=30)

df = pd.DataFrame({
    'Date': date_range,
    'Temperature' : temperature,
    'Humidity': humidity
})

# Step 2: Plot using dual y-axes
fig,ax1 = plt.subplots(figsize=(10,6))

# Plot Temperature on left y-axis
ax1.plot(df['Date'],df['Temperature'],color='red',label='Temperature (°C)')
ax1.set_ylabel('Temperature (°C)',color='red')
ax1.tick_params(axis = 'y' , labelcolor = 'red')

# Step 3: Create second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'],df["Humidity"],color='blue',label='Humidity (%)')
ax2.set_ylabel('Humidity (%)' , color='blue')
ax2.tick_params(axis='y',labelcolor='blue')

# Step 4: Final touches
plt.title('Temperature and Humidity Over Time')
ax1.set_xlabel('Date')
fig.tight_layout()
plt.show()
```

15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following  
tasks using Matplotlib:

 a) Plot a histogram of the data with 30 bins.

 b) Overlay a line plot representing the normal distribution's probability density function (PDF).

 c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.
 
 d) Set the title of the plot as 'Histogram with PDF Overlay'.




16.  Set the title of the plot as 'Histogram with PDF Overlay'.


```python
from scipy.stats import norm

# Step 1: Generate 1000 random samples from a normal distribution
data = np.random.randn(100)  # mean=0, std=1 by default

# Step 2: Create a histogram with 30 bins and density=True to normalize
count,bins,ignored = plt.hist(data,bins=30,density=True,alpha=0.6,color='skyblue',label='Histogram')

# Step 3: Calculate the PDF values using sample mean and std
pdf = norm.pdf(bins,loc=np.mean(data),scale=np.std(data))

# Step 4: Plot the PDF curve
plt.plot(bins,pdf,'-r',linewidth=2,label='Normal PDF')

# Step 5: Label axes and add title
plt.xlabel('Value')
plt.ylabel('Frequency/Probablity')
plt.title('Histogram with PDF Overlay')

# Step 6: Add legend
plt.legend()

plt.title('Histogram with PDF Overlay')

# Step 7: Display the plot
plt.show()
```

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the 
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import seaborn as sns

# Generate two random arrays of 100 float values each between -10 and 10
np.random.seed(42)
x = np.random.uniform(-10,10,100)
y = np.random.uniform(-10,10,100)

# Determine quadrant for each point
def get_quadrant(x,y):
    if x>0 and y>0:
        return 'Quadranr I'
    elif x<0 and y>0:
        return 'Quadrant II'
    
    elif x<0 and y<0:
        return 'Quadrant III'
    elif x>0 and y<0:
        return 'Quadrant IV'
    else:
        return 'On Axis'
    
    # Create DataFrame
df = pd.DataFrame({'X':x,'Y':y})
df['Quadrant'] = df.apply(lambda row: get_quadrant(row['X'],row['Y']),axis=1)

# Create scatter plot using Seaborn
plt.figure(figsize=(10,6))
sns.scatterplot(data=df , x='X',y='Y',hue='Quadrant',palette='Set1',s=100,edgecolor='black')

# Customize plot
plt.axhline(0,color='grey',linestyle='--',linewidth=1)
plt.axvline(0,color='grey',linestyle='--',linewidth=1)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.grid(True)
plt.tight_layout()
plt.show()
```
